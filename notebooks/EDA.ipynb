## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
df.head()
```

```python
df.info()
```

### Data Cleaning

```python
# missing values
df[df.isnull().any(axis=1)]
```

### The datastet is converted into two region ffrom 122th index, creating a new column based on region
#### 1: "Bejaia region dataset"
#### 2: "Sidi-Bel Abbes Region dataset"

##### adding new column of region

```python
df.loc[:122, 'Region']=0 # for Bejaia region
df.loc[122:, 'Region']=1 # for Sidi-Bel Abbes region
```

```python
df.Region.info() # it is in float, need to convert into int
```

```python
df.Region = df.Region.astype(int)
```

```python
df.isnull().sum()
```

```python
# Removing the null values

df=df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] # need to remove
```

```python
df= df.drop(122).reset_index(drop= True)
```

```python
df.columns  # spaces in column name
```

```python
# fixing spaces in column name
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

```python
# changing the required columns as integer
columns_list= ['day', 'month', 'year', 'Temperature', 'RH', 'Ws']

for i in columns_list:
    df[i] = df[i].astype(int)
```

```python
df.info()
```

```python
# changing the other columns to float datatype
object_features= [features for features in df.columns if df[features].dtype=='O']

for i in object_features:
    if i!= 'Classes':
        df[i]= df[i].astype(float)

df.info()
```

```python
df.describe()
```

```python
# lets save the clean dataset
df.to_csv('Algerian_forest_fires_dataset_CLEAN.csv', index=False)
```

```python
# percentage for Classes column

class_percentage=df_copy.Classes.value_counts(normalize=True) * 100
```

```python
# monthly fire analysis

df_temp = df.loc[df['Region']==0]
plt.figure(figsize= (8, 6))
sns.set_style('darkgrid')
sns.countplot(data= df_temp,
            x='month',
            hue='Classes',
            palette  = ['skyblue', 'orange'])
plt.title('Fire analysis Bejaia region', color= 'teal', weight = 'bold', fontsize= 16, loc= 'left')
plt.xlabel('months', color = 'red')
plt.ylabel('Number of fires', color = 'red')
plt.show()


# most fire in Auguest month
# less fire is in September month
```
